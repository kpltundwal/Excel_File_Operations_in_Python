#  8. Excel File Operations using Python (Pandas)

# Import Libraries

```python
import pandas as pd
```

# Read the Excel File 

```python
# Read the Excel file
df = pd.read_excel("student_data.xlsx")

# Display the first 5 rows of the dataset
df.head()
```

#  Update Marks for a Specific Student

```python
# Update marks for the student with Student_ID 105
df.loc[df['Student_ID'] == 105, 'Marks'] = 95

# Save the updated data to the same file
df.to_excel("student_data.xlsx", index=False)

# Print confirmation
print("Marks updated for Student_ID 105.")
```

#  Add a New Student Record

```python
# Create a new student record
new_student = {
    "Student_ID": 201,
    "Name": "Karan",
    "Age": 23,
    "Grade": "B",
    "Marks": 78
}

# Append the new record to the DataFrame
df = pd.concat([df, pd.DataFrame([new_student])], ignore_index=True)

# Save the updated data to the same file
df.to_excel("student_data.xlsx", index=False)

# Print confirmation
print("New student record added.")
```

# Delete a Column and a Row

```python
# Delete 'Grade' column
df.drop('Grade', axis=1, inplace=True)

# Delete the row with Student_ID 104
df = df[df['Student_ID'] != 104]

# Save the updated data to the same file
df.to_excel("student_data.xlsx", index=False)

# Print confirmation
print("Deleted 'Grade' column and Student_ID 104 record.")
```

#  Save the Updated Data

```python
# Save the updated DataFrame back to Excel
df.to_excel("student_data.xlsx", index=False)

# Print confirmation
print("Updated data saved to student_data.xlsx")
```
